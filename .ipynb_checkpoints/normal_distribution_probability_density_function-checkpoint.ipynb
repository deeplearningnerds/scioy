## Import Packages

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

## Probability density function of the Normal Distribution

We use the probability density function of the normal distribution.


Let's define our x values

```python
x = np.arange(-7, 7, 0.1)

print(x)
```

### Example 1 (Standard Normal Distribution): Mean = 0, Standard Deviation = 1

```python
mean = 0
standard_deviation = 1

y1 = norm.pdf(x, mean, standard_deviation)
```

Let's visualize our data with Matplotlib

```python
plt.plot(x, y1, color='b')
plt.xlim(-7, 7)
plt.show() 
```

### Example 2: Mean = 0, Standard Deviation = 0,5

```python
mean = 0
standard_deviation = 0.5

y2 = norm.pdf(x, mean, standard_deviation)
```

Let's visualize our data with Matplotlib

```python
plt.plot(x, y2, color='r')
plt.xlim(-7, 7)
plt.show() 
```

### Example 3: Mean = 1, Standard Deviation = 2

```python
mean = 1
standard_deviation = 2

y3 = norm.pdf(x, mean, standard_deviation)
```

Let's visualize our data with Matplotlib

```python
plt.plot(x, y3, color='g')
plt.xlim(-7, 7)
plt.show() 
```

### Comparison

```python
#Set figure size
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(6)

#plot 1
plt.plot(x, y1, color='b')

#plot 2
plt.plot(x, y2, color='r')

#plot 3
plt.plot(x, y3, color='g')

plt.tight_layout()
plt.show()
```
